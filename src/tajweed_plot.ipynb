```python
df_all = pd.read_csv('../data/tajweed_groups_30.csv')
df_all[df_all.columns.to_list()] = df_all[df_all.columns.to_list()].replace(0, np.nan)
df_all.head()
```

```python
df_exclude = pd.read_csv('../data/tajweed_groups_30_exclude.csv')
df_exclude[df_exclude.columns.to_list()] = df_exclude[df_exclude.columns.to_list()].replace(0, np.nan)
df_exclude.head()
```

```python
df_restrict = pd.read_csv('../data/tajweed_groups_10_restrict.csv')
df_restrict[df_restrict.columns.to_list()] = df_restrict[df_restrict.columns.to_list()].replace(0, np.nan)
df_restrict.head()
```

```python
df_all.columns.to_list()
```

```python
def plot_rules(df, cols):
    fig = px.scatter(df,
                     x='qindex',
                     y=cols,
                     labels=dict(chunk='chunks of words',),
                     title='Distribution of rules')
    fig.update_traces(marker_size=4)
    fig.update_layout(xaxis={'range':[-.4, df.shape[0]]})
    #fig.show()
    fig.show('png', scale=1, width=1600, height=800)
```

```python
plot_rules(df_restrict, df_restrict.columns.to_list())
```

```python
COLS = ['MIN-W_I', 'MIN-Y_I']

plot_rules(df_all, COLS)
```
